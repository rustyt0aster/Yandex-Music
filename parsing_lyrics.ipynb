```python
import pandas as pd
import numpy as np

from tqdm import tqdm
```

```python
from lyricsgenius import Genius
# зарегался и получил токен
#TOKEN = 'hpmEEUETx67lfMJmUJzx0PcPReS3Q5CAoyQUUyZ6O3Trmb7L9GW0625yMLHr1Wz3'
TOKEN = 'nLVmcm5npOE57Rfciq1X-DwLwpQ5ajoC-q5TY1NqXHwHcbXn4WkJppVooizuNkOj'
genius = Genius(TOKEN)
```

```python
def fetch_lyrics(title, artist):
     try:
         song = genius.search_song(title, artist)
         if song is not None:
             return song.lyrics
     except Exception as e:
         print(f"Error fetching lyrics for '{title}' by '{artist}': {str(e)}")
     return None
```

**Проба**

```python
data_lyrics = pd.read_csv('data_artist_song.csv', sep='\t')
```

```python
data_lyrics.head()
```

## Поиск текстов к песням

```python
for i in tqdm(range(len(data_lyrics))):
    data_lyrics.loc[i,'data_lyrics'] = fetch_lyrics(data_lyrics.loc[i,'title'], data_lyrics.loc[i,'artist']);
```
